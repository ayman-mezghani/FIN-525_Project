```python
data_path = 'data/5m/'
```

```python
df = pd.read_parquet(data_path + 'data.parquet')
df
```

```python
df.dropna(how='all', inplace=True)
df
```

```python
df.dropna(how='all', inplace=True, axis=1)
df
```

```python
df = df.stack(level=[0]).reset_index(level=[1]).rename(columns={'level_1': 'Ticker'})
df
```

```python
def preprocess(df):
    df['Log_Ret'] = np.log(df['Close']).diff()
    return df
```

```python
df['Log_Ret'] = df.groupby('Ticker').apply(lambda df: np.log(df['Close']).diff()).swaplevel().sort_index().values
```
